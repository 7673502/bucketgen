# BucketGen Demonstration
This notebook serves as a demonstration on how to use BucketGen.

**Note that this notebook requires having the [CIFAR-10](https://www.kaggle.com/datasets/swaroopkml/cifar10-pngs-in-folders) and [MNIST](https://www.kaggle.com/datasets/alexanderyyy/mnist-png) digits PNG dataset from Kaggle.**

```python
import bucketgen as bg
import matplotlib.pyplot as plt 
```

## A Quick Example

```python
dl = bg.DataLoader('mnist_png/train/0')
dl.images.shape
```

```python
generator = bg.Generator(dl.images)
```

```python
sample = generator.sample(show_result=True)
```

## MNIST

```python
dls = {}

for i in range(10):
    print(f'{i}: ', end='')
    dls[i] = bg.DataLoader(f'mnist_png/train/{i}')
```

```python
generators = {}

for i in range(10):
    print(f"now building {i} class generator")
    generators[i] = bg.Generator(dls[i].images)
```

```python
classes = ['zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']

plt.figure(figsize=(10, 4.5))

for idx, i in enumerate(classes):
    sample = generators[idx].sample()
    ax = plt.subplot(2, 5, idx + 1)
    ax.imshow(sample)
    ax.axis('off') 
    ax.set_title(i)
    
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
plt.show()
```

## CIFAR-10
The generator, as expected, does not produce an interpretable result when using the CIFAR-10 dataset.

```python
classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

dls = {}

for i in classes:
    print(f'{i}: ', end='')
    dls[i] = bg.DataLoader(f'cifar10/train/{i}')
```

```python
generators = {}

for i in classes:
    print(f"now building {i} class generator")
    generators[i] = bg.Generator(dls[i].images)
```

```python
plt.figure(figsize=(10, 4.5))

for idx, i in enumerate(classes):
    sample = generators[i].sample()
    ax = plt.subplot(2, 5, idx + 1)
    ax.imshow(sample)
    ax.axis('off') 
    ax.set_title(i) 
    
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
plt.show()
```
